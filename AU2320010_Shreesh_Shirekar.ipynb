# Reverse a Linked List

```python
class Node: # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None
        
class linked_list: #initializing linked list
    def __init__(self):
        self.head = None
        
    def insert(self, data): # insert function for inserting the nodes
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
    

    def reverse(self): # reverses a function
        prev = None # initializes previous pointer
        current = self.head # sets the current pointer to the head
        
        while current: # while current is not None
            next_node = current.next # goes to next node
            current.next = prev # reverse the pointer
            prev = current # move the prev pointer to the current node
            current = next_node # then move to next node
            
        self.head = prev # update the last pointer as the head
        
    def transverse(self): # transverse function to display the linked list
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
        
llist = linked_list()

llist.insert(1)
llist.insert(2)
llist.insert(3)
llist.insert(4)
llist.insert(5)

llist.transverse()
llist.reverse()
print("Reversing the list....")
llist.transverse()
```

# Reverse a Doubly Linked List

```python
class Node: # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class doubly_linked_list: #initializing linked list
    def __init__(self):
        self.head = None
        
    def insert(self, data): # insert function for inserting the nodes
        new_node = Node(data)
        new_node.next = self.head
        if self.head:
            self.head.prev = new_node
        self.head = new_node
    

    def reverse(self): # reverses a function
        prev = None # initializes previous pointer
        current = self.head # sets the current pointer to the head
        
        while current: # while current is not None
            next_node = current.next # goes to next node
            current.next = prev # reverse the pointer
            current.prev = next_node # reverse the previous pointer
            prev = current # move the prev pointer to the current node
            current = next_node # then move to next node
            
        self.head = prev # update the last pointer as the head
        
    def transverse(self): # transverse function to display the linked list
        current = self.head
        while current:
            print(current.data, end=" <-> ")
            current = current.next
        print("None")
        
dll = doubly_linked_list()
dll.insert(1)
dll.insert(2)
dll.insert(3)
dll.insert(4)
dll.insert(5)

dll.transverse()
print("Reversing the list....")
dll.reverse()
dll.transverse()
```

# Rotate a Linked List

```python
class Node: # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None
        
class linked_list: #initializing linked list
    def __init__(self):
        self.head = None
        
    def insert(self, data): # insert function for inserting the nodes
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
    def rotate(self, k):
        if not self.head or k == 0: # if the list is empty or if the k = 0, no rotation
            return
        
        current = self.head # finding the length of the list
        length = 1 # set counter to one and iteratively traverse the list and count
        while current.next:
            current = current.next
            length += 1
        
        k = k % length # if k is larger than length of node, rotate only the remainder steps
        if k == 0:
            return
        
        updated_last = self.head # iterates and then finds the k-1 node for the tails
        for _ in range(k - 1):
            updated_last = updated_last.next
            
        updated_head = updated_last.next # this assigns the new head from the last tail
        updated_last.next = None # removes pointer to make the last node the tail
        current.next = self.head # connects the old last node to the old head
        
        self.head = updated_head # updates the head to the new head
        
    def traverse(self): # transverse function to display the linked list
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
        
        # Example usage:
sll = linked_list()
sll.insert(1)
sll.insert(2)
sll.insert(3)
sll.insert(4)
sll.insert(5) 


sll.traverse()

sll.rotate(2)

print("List after rotating:")
sll.traverse()        
```

# Nth Node from End of Linked List

```python
class Node: # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None

class linked_list: # initializing linked list
    def __init__(self):
        self.head = None

    def insert(self, data): # insert function for inserting the nodes
        new_node = Node(data) 
        new_node.next = self.head 
        self.head = new_node # assigns the new node as the head

    def nth_from_end(self, n): # function to find nth node from end
        first = self.head # initialize two pointers: first and second
        second = self.head 

        # move the first pointer n steps ahead in the list
        for _ in range(n):
            if not first: # if n is greater than the length of the list
                print(f"List has fewer than {n} nodes.")
                return
            first = first.next

        # move both pointers until the first pointer reaches the end
        while first:
            first = first.next
            second = second.next # second pointer moves with first pointer

        print(f"The {n}th node from the end is: {second.data}")

    def traverse(self): # traverse function to display the linked list
        current = self.head
        while current:
            print(current.data, end=" -> ") # prints data in each node
            current = current.next
        print("None") # indicates the end of the list


sll = linked_list()
sll.insert(1)
sll.insert(2)
sll.insert(3)
sll.insert(4)
sll.insert(5)
sll.traverse()

sll.nth_from_end(2)
```

# Delete Last Occurrence of an Item from Linked List

```python
class Node:  # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None

class linked_list:  # initializing linked list
    def __init__(self):
        self.head = None

    def insert(self, data):  # insert function for inserting the nodes
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node  # assigns the new node as the head

    def delete_last_occurrence(self, x):  # function to delete last occurrence of x
        current = self.head
        last_occurrence = None  # store the previous node of last occurrence of x
        prev = None  # keep track of the previous node during traversal

        # Traverse the entire list to find the last occurrence of x
        while current:
            if current.data == x:  # when x is found, store the current node
                last_occurrence = prev
            prev = current  # move previous pointer
            current = current.next  # move current pointer

        if last_occurrence is None:  # x not found in the list
            print(f"{x} not found in the list.")
            return

        # If the last occurrence is found, delete it by skipping the node
        if last_occurrence.next:  # skip the node containing the last occurrence of x
            last_occurrence.next = last_occurrence.next.next
        else:  # edge case: if the node to delete is the head
            self.head = self.head.next

    def traverse(self):  # traverse function to display the linked list
        current = self.head
        while current:
            print(current.data, end=" -> ")  # prints data in each node
            current = current.next
        print("None")  # indicates the end of the list

sll = linked_list()
sll.insert(1)
sll.insert(2)
sll.insert(3)
sll.insert(2)
sll.insert(4)
sll.insert(2)


sll.traverse()

sll.delete_last_occurrence(2)

print("List after deleting the last occurrence of 2:")
sll.traverse()
```

# Delete Middle of a Linked List

```python
class Node:  # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None

class linked_list:  # initializing linked list
    def __init__(self):
        self.head = None

    def insert(self, data):  # insert function for inserting the nodes
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node  # assigns the new node as the head

    def delete_middle(self):  # function to delete the middle node
        if not self.head or not self.head.next:  # if list has 0 or 1 node
            self.head = None  # set the head to None
            return

        slow = self.head  # initialize slow and fast pointers
        fast = self.head
        prev = None  # store the previous node of the slow pointer

        # Move fast by 2 steps and slow by 1 step until fast reaches the end
        while fast and fast.next:
            prev = slow  # update the previous node
            slow = slow.next  # move slow by 1 step
            fast = fast.next.next  # move fast by 2 steps

        # Delete the middle node by skipping it
        if prev:
            prev.next = slow.next

    def traverse(self):  # traverse function to display the linked list
        current = self.head
        while current:
            print(current.data, end=" -> ")  # prints data in each node
            current = current.next
        print("None")  # indicates the end of the list

sll = linked_list()
sll.insert(1)
sll.insert(2)
sll.insert(3)
sll.insert(4)
sll.insert(5) 


sll.traverse()

sll.delete_middle() 

print("List after deleting the middle node:")
sll.traverse()
```

# Remove Duplicate Elements from Sorted Linked List

```python
class Node:  # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None

class linked_list:  # initializing linked list
    def __init__(self):
        self.head = None

    def insert(self, data):  # insert function for inserting nodes at the head
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def remove_duplicates(self):  # function to remove duplicates from sorted list
        current = self.head

        while current and current.next:  # traverse until the end of the list
            if current.data == current.next.data:  # if duplicate found
                current.next = current.next.next  # skip the duplicate node
            else:
                current = current.next  # move to the next distinct node

    def traverse(self):  # traverse function to display the linked list
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

sll = linked_list()
sll.insert(3)
sll.insert(3)
sll.insert(2)
sll.insert(2)
sll.insert(1)

sll.traverse()

sll.remove_duplicates()

print("List after removing duplicates:")
sll.traverse()
```

# Detect Loop in Linked List

```python
class Node:  # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None

class linked_list:  # initializing linked list
    def __init__(self):
        self.head = None

    def insert(self, data):  # insert function for inserting nodes at the head
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def detect_loop(self):  # function to detect loop in the linked list
        slow = self.head  # initialize both pointers to the head
        fast = self.head

        while fast and fast.next:  # traverse until the fast pointer reaches the end
            slow = slow.next  # move slow pointer by 1 step
            fast = fast.next.next  # move fast pointer by 2 steps

            if slow == fast:  # if both pointers meet, there is a loop
                return True

        return False  # if no loop is found

    def traverse(self):  # traverse function to display the linked list
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

sll = linked_list()
sll.insert(1)
sll.insert(2)
sll.insert(3)
sll.insert(4)
sll.insert(5)

# Creating a loop for testing: Point the last node (1) to the node with value 3
sll.head.next.next.next.next.next = sll.head.next.next

# Check for loop
if sll.detect_loop():
    print("Loop detected.")
else:
    print("No loop detected.")
```

# Merge a Linked List into Another Linked List at Alternate Positions

```python
class Node:  # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None

class linked_list:  # initializing linked list
    def __init__(self):
        self.head = None

    def insert(self, data):  # function to insert a new node at the beginning
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node  # assigns the new node as the head

    def merge_alternate(self, other):  # function for merging the lists alternately
        first = self.head  # set first pointer to the head of the first list
        second = other.head  # set second pointer to the head of the second list

        # Loop until either list is fully traversed
        while first and second:
            first_next = first.next  # store the next node of the first list
            second_next = second.next  # store the next node of the second list

            first.next = second  # link the current node of the first list to the current node of the second list
            second.next = first_next  # link the current node of the second list to the next node of the first list

            first = first_next  # move first pointer to the next node in the first list
            second = second_next  # move second pointer to the next node in the second list

        other.head = second  # update head of second list to remaining nodes, if any

    def traverse(self):  # function to display the linked list
        current = self.head  # start from the head of the list
        while current:  # traverse until the end of the list
            print(current.data, end=" -> ")  # print data in each node
            current = current.next  # move to the next node
        print("None")  # indicates the end of the list


ll1 = linked_list()
ll2 = linked_list()

ll1.insert(3)
ll1.insert(2)
ll1.insert(1)

ll2.insert(6)
ll2.insert(5)
ll2.insert(4)

print("First Linked List:")
ll1.traverse()

print("Second Linked List:")
ll2.traverse()

ll1.merge_alternate(ll2)

print("Merged Linked List:")
ll1.traverse()

print("Remaining nodes in the second list (if any):")
ll2.traverse()
```

# Implement a Queue using Stacks

```python
class queue_stacks:
    def __init__(self):
        self.stack1 = []  # stack1 for enqueue operations
        self.stack2 = []  # stack2 for dequeue operations

    def enqueue(self, data):
        self.stack1.append(data)  # pushing element onto Stack1

    def dequeue(self):
        if not self.stack2:  # if stack2 is empty, move elements from stack1 to stack2
            while self.stack1:
                self.stack2.append(self.stack1.pop())

        if not self.stack2:  # if both stacks are empty, queue is empty
            print("Queue is empty.")
            return None
        
        return self.stack2.pop()  # popping from stack2

q = queue_stacks()

q.enqueue(1)
q.enqueue(2)
q.enqueue(3)

print(q.dequeue())
print(q.dequeue())

q.enqueue(4)
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
```

# Circular Queue Implementation

```python
class circular_queue:  # initializing the circular queue class
    def __init__(self, size):  # constructor to initialize queue size and attributes
        self.size = size
        self.queue = [None] * size  # creates an array of given size initialized with None
        self.front = -1  # indicates the front of the queue, -1 when queue is empty
        self.rear = -1   # indicates the rear of the queue, -1 when queue is empty

    def isFull(self):  # checks if the queue is full
        return (self.rear + 1) % self.size == self.front  # returns True if rear is one step behind front

    def isEmpty(self):  # checks if the queue is empty
        return self.front == -1  # returns True if front is -1

    def enqueue(self, value):  # function to add an element to the queue
        if self.isFull():  # if the queue is full
            print("Queue is full")  # print queue is full and exit the function
            return
        if self.isEmpty():  # if the queue is empty
            self.front = 0  # initialize front to 0 for the first element
        self.rear = (self.rear + 1) % self.size  # update rear position (wraps around if needed)
        self.queue[self.rear] = value  # assign the value to the rear position
        print(f"Enqueued: {value}")  # print the enqueued value

    def dequeue(self):  # function to remove an element from the queue
        if self.isEmpty():  # if the queue is empty
            print("Queue is empty")  # print queue is empty and exit the function
            return None
        value = self.queue[self.front]  # store the front element
        if self.front == self.rear:  # if there’s only one element left
            self.front = -1  # reset front and rear to -1 indicating an empty queue
            self.rear = -1
        else:
            self.front = (self.front + 1) % self.size  # update front position (wraps around if needed)
        print(f"Dequeued: {value}")  # print the dequeued value
        return value  # return the dequeued element

    def display(self):  # function to display elements in the queue
        if self.isEmpty():  # if the queue is empty
            print("Queue is empty")  # print queue is empty and exit the function
            return
        print("Queue elements:", end=" ")
        i = self.front  # start from the front
        while True:
            print(self.queue[i], end=" ")  # print each element in the queue
            if i == self.rear:  # stop when we reach the rear
                break
            i = (i + 1) % self.size  # move to the next position (wraps around if needed)
        print()  # new line after printing all elements


# Initialize a circular queue with a capacity of 5
cq = circular_queue(5)

# Enqueue elements
cq.enqueue(10)
cq.enqueue(20)
cq.enqueue(30)
cq.enqueue(40)
cq.enqueue(50)  

cq.display()  

# Dequeue a few elements
cq.dequeue()  
cq.dequeue()  

cq.display()  

# Enqueue more elements to wrap around
cq.enqueue(60)
cq.enqueue(70)

cq.display()  
```

# Reverse the First N Elements of a Queue

```python
class Node:  # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None

class queue:  # initializing the linked list queue
    def __init__(self):
        self.front = None  # start with an empty queue
        self.rear = None

    def enqueue(self, x):  # function to add an element to the queue
        new_node = Node(x)
        if self.isEmpty():  # if the queue is empty
            self.front = self.rear = new_node  # set both front and rear to the new node
        else:
            self.rear.next = new_node  # link the new node to the end of the queue
            self.rear = new_node  # update rear to the new node

    def dequeue(self):  # function to remove an element from the queue
        if self.isEmpty():  # if the queue is empty
            print("Queue is empty")
            return None
        value = self.front.data  # store the data to return
        self.front = self.front.next  # move front to the next node
        if self.front is None:  # if the queue is now empty
            self.rear = None  # reset rear to None as well
        print(f"Dequeue: {value}")  # print the dequeued value
        return value

    def peek(self):  # function to return the front element without removing it
        if self.isEmpty():  # if the queue is empty
            print("Queue is empty")
            return None
        print(f"Peek: {self.front.data}")  # print the front element
        return self.front.data

    def isEmpty(self):  # function to check if the queue is empty
        return self.front is None  # returns True if front is None

    def display(self):  # function to display elements in the queue
        if self.isEmpty():  # if the queue is empty
            print("Queue is empty")
            return
        current = self.front  # start from the front of the queue
        print("Queue elements:", end=" ")
        while current:
            print(current.data, end=" -> ")  # print data in each node
            current = current.next  # move to the next node
        print("None")  # indicates the end of the queue

    def reverse_first_n(self, n):  # function to reverse the first N elements
        if n <= 0 or self.isEmpty():  # if n is non-positive or queue is empty
            return

        stack = []  # use a stack to reverse the first N elements
        # Dequeue first N elements and push them onto the stack
        for _ in range(n):
            if self.isEmpty():  # if fewer than N elements in the queue
                break
            stack.append(self.dequeue())  # dequeue and push to the stack

        # Re-enqueue the elements from the stack (reversed order)
        while stack:
            self.enqueue(stack.pop())  # pop from the stack and enqueue back

        # Move the remaining elements to the back to maintain original order
        size = 0
        current = self.front
        while current:
            size += 1
            current = current.next
        
        for _ in range(size - n):  # dequeue remaining elements and enqueue back
            self.enqueue(self.dequeue())


q = queue()
q.enqueue(5)
q.enqueue(10)
q.enqueue(15)
q.enqueue(20)
q.enqueue(25)

print("Original Queue:")
q.display()

q.reverse_first_n(3)
print("Queue after reversing first 3 elements:")
q.display()

q.dequeue()
q.peek()
q.isEmpty()
```

# Check if a Queue is a Palindrome

```python
class Node:  # initializing the node class
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:  # initializing the linked list queue
    def __init__(self):
        self.front = None  # start with an empty queue
        self.rear = None
        self.count = 0  # to keep track of the queue size

    def enqueue(self, x):  # function to add an element to the queue
        new_node = Node(x)
        if self.isEmpty():  # if the queue is empty
            self.front = self.rear = new_node  # set both front and rear to the new node
        else:
            self.rear.next = new_node  # link the new node to the end of the queue
            self.rear = new_node  # update rear to the new node
        self.count += 1  # increment queue size

    def dequeue(self):  # function to remove an element from the queue
        if self.isEmpty():  # if the queue is empty
            return None
        value = self.front.data  # store the data to return
        self.front = self.front.next  # move front to the next node
        if self.front is None:  # if the queue is now empty
            self.rear = None  # reset rear to None as well
        self.count -= 1  # decrement queue size
        return value

    def isEmpty(self):  # function to check if the queue is empty
        return self.front is None  # returns True if front is None

    def size(self):  # function to return the size of the queue
        return self.count  # returns the count of elements in the queue

    def display(self):  # function to display elements in the queue
        if self.isEmpty():  # if the queue is empty
            print("Queue is empty")
            return
        current = self.front  # start from the front of the queue
        print("Queue elements:", end=" ")
        while current:
            print(current.data, end=" -> ")  # print data in each node
            current = current.next  # move to the next node
        print("None")  # indicates the end of the queue

def is_palindrome(q):
    stack = []  # auxiliary stack to hold the elements in reverse order
    n = q.size()  # get the initial size of the queue

    # transfer elements to stack and re-enqueue to maintain order
    for _ in range(n):
        elem = q.dequeue()  # dequeue element from queue
        stack.append(elem)  # push element onto stack
        q.enqueue(elem)  # re-enqueue element to preserve original order

    # check if queue is a palindrome by comparing with stack
    is_palindrome = True
    for _ in range(n):
        elem = q.dequeue()  # dequeue element from queue
        if elem != stack.pop():  # compare with top of stack
            is_palindrome = False
        q.enqueue(elem)  # re-enqueue element to restore original order

    return is_palindrome

q1 = Queue()
for elem in [1, 2, 3, 2, 1]:
    q1.enqueue(elem)

print("Is Palindrome:", is_palindrome(q1))
q2 = Queue()
for elem in [1, 2, 3, 4, 5]:
    q2.enqueue(elem)

print("Is Palindrome:", is_palindrome(q2))
```
